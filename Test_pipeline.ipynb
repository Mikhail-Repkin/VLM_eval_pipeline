# Тестирование моделей

## Сценарий тестирования

**Сценарий тестирования:**    
- один промт
- температура тест-моделей: 0.1
- 5 запусков генерации х 10 картинок = 50 генераций на 1 модель

**Технические статистики:**
- Общее время генерации (сек.)
- Средняя длина ответа (знаки)
- Скорость генерации — кол-во знаков в секунду
- Количество грамматических ошибок
- Количество некириллических и не англоязычных символов
- Доля англоязычных символов

**Автоматические метрики:**
* Перплексия — насколько хорошо модель предсказывает последовательности слов (чем ниже, тем лучше).
* BLEURT — метрика, основанная на предобученных трансформерах, которая оценивает качество текста по его смысловому сходству с эталонным текстом. BLEURT учитывает синонимы и парафразы, лексическую и семантическую близость, относительную важность различных слов. 0 означает случайный результат, а 1 - идеальный.

**Субъективная (судейская) оценка:**
* На основе промпта для LLM судья ставит оценки по десятибалльной шкале и подробно объясняет, почему.
* Учитывается анонимность ответов. Судья не знает и не учитывает, от какой модели приходит ответ.
* Ответы рандомно меняются местами, чтобы избежать смещения в пользу какой-то одной модели (Position bias - первый ответ выбирается чаще).
* В качестве эталонной модели (референс) выступает gpt-4o
* В качестве модели-судьи (LLM-as-a-Judge) выступает gpt-4o-mini
* Также оценивается доля соответствия тестируемой модели к эталонной, где эталонная модель имеет долю соответсвия 1.0 (доля соответствия - отношение судейских оценок).

Модель разворачивается с помощью фреймворка vLLM раз как сервер, реализующий протокол OpenAI API. По умолчанию сервер запускается по адресу http://localhost:8000. Сервер использует предопределенный [шаблон чата](https://github.com/vllm-project/vllm/blob/main/docs/source/serving/openai_compatible_server.md#chat-template), хранящийся в токенизаторе.

К этому серверу можно обращаться в том же формате, что и к OpenAI API. Например, чтобы получить список моделей:
curl http://localhost:8000/v1/models

## Тест: Qwen/Qwen2-VL-7B-Instruct

```python
MODEL_PATH = "Qwen/Qwen2-VL-7B-Instruct"
NUM_VARIANTS = 5  # кол-во генераций

temperatures = [0.1]

# Запуск сервера, генерация текстов и сбор статистики
generator = test_eval.ImageDescriptionGenerator(MODEL_PATH,
                                      DEFAULT_SYSTEM_MESSAGE,
                                      USER_MESSAGE)
generator.start_server()

results = generator.generate_descriptions_for_images(image_paths,
                                                     temperatures,
                                                     base64=False,
                                                     num_variants=NUM_VARIANTS)

# Подсчет орфографических ошибок
spell_checker = tech_metrics_eval.SpellChecker()
texts = list(results['Generated Text'])
spell_errors = spell_checker.analyze_texts(texts)

# Подсчет ошибок языка
lang_errors = tech_metrics_eval.count_foreign_letters_in_texts(texts)

# Вычисление Перплексии текстов
perplexity = tech_metrics_eval.calculate_perplexity_for_texts(texts)

# Объединение и вывод статистик
model_name_column = pd.DataFrame({"model_name": [MODEL_PATH.split('/')[-1]] * results.shape[0]})
final_stats = pd.concat([model_name_column, results, spell_errors, lang_errors, perplexity], axis=1)
display(final_stats)

# Сохраняем в Excel
final_stats.to_excel(f"Test_{MODEL_PATH.split('/')[-1]}_stats.xlsx", index=False)
```

Определим модели, которые сваливаются в генерацию на английском языке.

```python
combined_data_copy = combined_data.copy()

# Функция для извлечения уникальных английских слов
def extract_unique_english_words(text):
    words = text.split()
    english_words = {word.strip(".,!?") for word in words if word.isascii() and word.isalpha()}
    return list(english_words)

combined_data_copy["unique_en_words"] = combined_data_copy["candidate"].apply(extract_unique_english_words)
combined_data_copy["unique_en_count"] = combined_data_copy["unique_en_words"].apply(len)

df_unique_en_words = combined_data_copy[["model_name", "unique_en_words", "Text Length (char)", "unique_en_count"]]

df_unique_en_words = df_unique_en_words.copy()
df_unique_en_words.loc[:,'EN chars ratio'] = (df_unique_en_words['unique_en_count'] / df_unique_en_words['Text Length (char)']).apply(lambda x: round(x, 3))
# df_unique_en_words = df_unique_en_words[df_unique_en_words["unique_english_words"].apply(len) > 0]

df_unique_en_words_pivot = df_unique_en_words.pivot_table(index="model_name",
                                                          values=["unique_en_words", "unique_en_count", "Text Length (char)"],
                                                          aggfunc={"unique_en_words": lambda x: list(set(word for sublist in x for word in sublist)),
                                                                   "Text Length (char)": "sum",
                                                                   "unique_en_count": "sum"})
df_unique_en_words_pivot["EN chars, %"] = (df_unique_en_words_pivot["unique_en_count"] / df_unique_en_words_pivot["Text Length (char)"]).apply(lambda x: round(x * 100, 1))
df_unique_en_words_pivot.reset_index(inplace=True)
df_unique_en_words_pivot
```

```python
JUDGE_DEFAULT_SYSTEM_MESSAGE = (
            "You are a neutral judge evaluating possible user questions generated from a given screenshot and a related message.\n"
            "You have an image and two different user questions that have been generated by two unknown models.\n"
            "You do not know which model produced which question. Your task is to fairly and strictly assess the quality of each question.\n"
            "Evaluate each question on a scale from 1 to 10, based on how well it is:\n"
            "- Clearly derived from the provided screenshot and message\n"
            "- Relevant to the information shown or described\n"
            "- Specific, helpful, and not vague\n"
            "- Formulated in a natural, user-friendly way\n\n"
            "After evaluating both questions, provide a detailed explanation of why you assigned these scores.\n"
            "Do not reveal or guess which model provided each question. Simply call them 'Answer A' and 'Answer B' in the order they are given.\n"
            "Finally, return your judgment in JSON format:\n"
            "{\n"
            "\"ModelA_score\": <rating for the first answer>,\n"
            "\"ModelB_score\": <rating for the second answer>,\n"
            "\"Explanation\": \"<detailed explanation>\"\n"
            "}"
            )

USER_MESSAGE = ""

print(JUDGE_DEFAULT_SYSTEM_MESSAGE)
```

```python
encoding = tiktoken.encoding_for_model("gpt-4o-mini")
token_count = len(encoding.encode(JUDGE_DEFAULT_SYSTEM_MESSAGE))

print("Number of tokens in the prompt:", token_count)
```

```python
JUDGE_MODEL_PATH = "gpt-4o-mini"  # Модель-судья

image_paths = {
    "test_1": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_1.jpg",
    "test_2": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_2.jpg",
    "test_3": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_3.jpg",
    "test_4": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_4.jpg",
    "test_5": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_5.jpg",
    "test_6": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_6.jpg",
    "test_7": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_7.jpg",
    "test_8": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_8.jpg",
    "test_9": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_9.jpg",
    "test_10": "https://raw.githubusercontent.com/Mikhail-Repkin/VLM_eval_pipeline/refs/heads/main/open_test_dataset/test_10.jpg",
}

# model_name | Image | candidate | reference
data_path = "candidates_references.xlsx"
output_path = "judgement_scores.xlsx"

# Инициализируем судейский класс
model_judge = judge_eval.ModelComparisonJudge(JUDGE_MODEL_PATH,
                                              gpt4_api_key,
                                              JUDGE_DEFAULT_SYSTEM_MESSAGE,
                                              USER_MESSAGE)

# Загрузка данных
if os.path.exists(output_path):
    final_data = pd.read_excel(output_path)
    processed_indices = set(final_data.index)
else:
    final_data = pd.DataFrame()
    processed_indices = set()

data = pd.read_excel(data_path)
results = []

try:
    with tqdm(total=len(data), desc="Evaluating") as pbar:
        for idx, row in data.iterrows():
            if idx in processed_indices:
                pbar.update(1)
                continue  # Пропускаем уже обработанные строки

            img_url = image_paths[row['Image']]
            answer_A = row['reference']  # эталон
            answer_B = row['candidate']  # кандидат
            print("img_url:", img_url)
            print("answer_A:", answer_A)
            print("answer_B:", answer_B)

            # Оценка ответов
            judgement = model_judge.judge_responses(img_url, answer_A, answer_B)
            print("Ответ судьи:", judgement)

            # Добавляем результаты
            results.append({
                "Reference_Score": judgement["Reference_score"],
                "Candidate_Score": judgement["Candidate_score"],
                "Judge Explanation": judgement["Explanation"]
            })

            # Сохраняем данные построчно
            temp_result = pd.DataFrame([{
                **row,  # Исходные данные
                "Reference_Score": judgement["Reference_score"],
                "Candidate_Score": judgement["Candidate_score"],
                "Judge Explanation": judgement["Explanation"]
            }])
            final_data = pd.concat([final_data, temp_result], axis=0)
            final_data.to_excel(output_path, index=False)
            print("Данные сохранены")
            print()
            pbar.update(1)

except Exception as e:
    print(f"Произошла ошибка: {e}")
    print("Данные были сохранены до момента ошибки.")

print("Процесс завершён. Результаты сохранены в judgement_scores.xlsx")
```

```python
judgement_scores_raw = pd.read_excel("judgement_scores.xlsx")

# Удаление выбросов в оценках
judgement_scores_clean = judgement_scores_raw[
    (judgement_scores_raw['Reference_Score'] >= 0) & (judgement_scores_raw['Reference_Score'] <= 10) &
    (judgement_scores_raw['Candidate_Score'] >= 0) & (judgement_scores_raw['Candidate_Score'] <= 10)
    ].copy()
print("Удалено выбросов:", len(judgement_scores_raw) - len(judgement_scores_clean))

# Добавление доли соответствия Candidate_Score к Reference_Score
judgement_scores_clean['match_ratio'] = judgement_scores_clean['Candidate_Score'] / judgement_scores_clean['Reference_Score']

pivot_table = judgement_scores_clean.groupby('model_name').agg(
    avg_Reference_Score=('Reference_Score', 'mean'),
    avg_Candidate_Score=('Candidate_Score', 'mean'),
    avg_match_ratio=('match_ratio', 'mean'),
    std_match_ratio=('match_ratio', 'std'),
    count=('match_ratio', 'count')
).reset_index()

pivot_table[['avg_Reference_Score', 'avg_Candidate_Score']] = pivot_table[
    ['avg_Reference_Score', 'avg_Candidate_Score']
].round(1)

pivot_table[['avg_match_ratio', 'std_match_ratio']] = pivot_table[
    ['avg_match_ratio', 'std_match_ratio']
].round(2)

pivot_table['std_err'] = pivot_table['std_match_ratio'] / (pivot_table['count'] ** 0.5)
pivot_table['conf_interval'] = 1.96 * pivot_table['std_err']

display(pivot_table)
```
